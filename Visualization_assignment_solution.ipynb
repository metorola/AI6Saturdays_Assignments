# Data Visualization Assignment - Solutions

```python
import numpy as np
import matplotlib.pyplot as plt

data_population = np.loadtxt("pop.txt", delimiter = ",")
data_year = np.loadtxt("year.txt", delimiter = ",")

print('Population:',data_population[-1])
print('Year:',data_year[-1])
```

```python
pop = np.c_[data_population]  #stack into columns
year = np.c_[data_year]

plt.figure(figsize=(10,10))   #size of the image
plt.plot(year,pop)
plt.xlabel('Year')            #labelling the axis
plt.ylabel('Population')
plt.xticks(np.arange(min(year), max(year)+1, 10.0))   #adjusting the elements on the x axis
plt.show()
```

```python
data_gdp = np.loadtxt("gdp_cap.txt", delimiter = ",")
data_lifexp = np.loadtxt("life_exp.txt", delimiter = ",")

print('GDP per capita:',data_gdp[-1])
print('Life Expectancy:',data_lifexp[-1])

gdp_cap = np.c_[data_gdp]
life_exp = np.c_[data_lifexp]

plt.plot(gdp_cap,life_exp)
plt.xlabel('GDP')            
plt.ylabel('Life Expectancy')
plt.show()
```

From the GDP - Life_Expectancy figure above, the lines appear scattered and looks like it makes no sense,but it appears that the highest GDPs do not have a low Life Expectancy.

```python
plt.scatter(gdp_cap,life_exp)
plt.xlabel('GDP')            
plt.ylabel('Life Expectancy')
plt.xscale('log')
plt.show()
```

WIth the scatter plot, its more clearer that the highest GDP correlates with the highest Life expectancy

```python
plt.scatter(gdp_cap, life_exp, edgecolor = 'black')

xlab = 'GDP per Capita [in USD]'
ylab = 'Life Expectancy [in years]'
title = 'World Development in 2007'

plt.xlabel(xlab)
plt.ylabel(ylab)
plt.title(title)

plt.xscale('log')
tick_val = [1000, 10000, 100000]
tick_lab = ['1k','10k','100k']
plt.xticks(tick_val,tick_lab)
plt.show()
```

```python
data_pop3 = np.loadtxt('pop3.txt', delimiter=',')

np_pop3 = np.array(data_pop3)
np_pop3 = np_pop3 * 4

plt.scatter(gdp_cap, life_exp, s = np_pop3, edgecolor = 'black', linewidth = 1.2)

plt.xscale('log') 
plt.xlabel('GDP per Capita [in USD]')
plt.ylabel('Life Expectancy [in years]')
plt.title('World Development in 2007')
plt.xticks([1000, 10000, 100000],['1k', '10k', '100k'])
plt.show()
```
